# Intro sur seaborn

C'est une srucouche à Matplotlib qui est adapté aux DataFrame et qui fait tout même la vaisselle. 

```python
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
sns.set(style="whitegrid", palette="muted")
```

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
sns.pairplot(iris, hue="species")
```

```python
iris2 = pd.melt(iris, "species", var_name="measurement")
sns.swarmplot(x="measurement", y="value", hue="species", data=iris2)
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
# Load the example tips dataset


# Draw a nested violinplot and split the violins for easier comparison
sns.violinplot(x="day", y="total_bill", hue="sex", data=tips, split=True,
               inner="quart", palette={"Male": "b", "Female": "y"})
sns.despine(left=True)
```

# Exercie 1

Sur le dataset du titanic

 * Avec la fonction sns.pairplot étudiez les variables "Fare","Age","Pclass"
 * Utilisez "survied pour la couleur 
 

# Solution 1

```python
passenger = pd.read_excel('./data/dataset_titanic.xls')
data = passenger[["Fare","Age","Pclass"]].copy()
data.loc[data['Age'].isnull(), 'Age'] = -10.
sns.pairplot(data=data, vars=['Age', 'Fare'], hue="Pclass", dropna=True)
```
